```python
#반복문(while)
#반복문을 통하여 누적을 구하는 예제
i = 0                #반복 카운터 i 0으로 초기화
sum = 0              #누적할 변수 sum 0으로 초기화
while i<100 :       #while문조건 : 0~100까지 실행(실행횟수는 100회)
    i+=1            #i는 1씩 증가( i ++ )
    sum += i        #sum은 누적하여 합해준다. 1+2+3+4+.........+99+100
print("sum=", sum)  #while반복문을 탈출하여 출력한다.
```

```python
#반복문 while 사례1
i = 0
sum = 0
while i<100 :
    i+=1           #index값을 먼저 올려주고 그거에 맞춰서 sum 값에 넣어준다.
    sum += i
print("sum=",sum)
```

```python
#반복문 while 사례2
i = 0
sum = 0
while i<100 :
    sum +=i       #0부터 시작해서 마지막에 100이 포함되지 않은 채 sum 출력
    i+=1
print("sum=",sum)
```

```python
#반복문 while 사례3
i = 0
sum = 0
while i<101 :
    sum += i     #이건 인덱스 자체가 100까지 포함되니까 기존과 동일하게 출력
    i+=1
print("sum=",sum)
```

```python
#반복문 (for)
#range 사용 사례 1 
#range(시작, 끝, 증가폭)
for i in range (10) :
    print(i, ",", end="")
```

```python
#반복문 (for)
#range 사용 사례 2
for i in range (1,10) :
    print(i, ",", end="")
```

```python
#반복문 (for)
#range 사용 사례 3
#range(시작, 끝, 증가폭)
for i in range (1, 10, 2) :            #증가폭이 2니까 2씩 올라감
    print(i, ",", end="")
```

```python
#반복문 (for)
#range 사용 사례 4
#range(시작, 끝, 증가폭)
for i in range (1, 10, 1) :            #10은 포함안되니까 총 9번 실행함
    print(i, ",", end="")
```

```python
#반복문 (for)
#range 사용 사례 5
#range(시작, 끝, 증가폭)
for i in range (10, 0, -1) :            #10은 포함안되니까 총 9번 실행함
    print(i, ",", end="")                #증가식이 -1이니까 1씩 감소함
```

```python
#제어문 예제 1
i = 0; sum = 0
while True:            #while 무한루프 돌리기
    i += 1; sum+=i      #index++; sum은 누적합
    if(i==100) : break
print("i=",i," sum=",sum)
```

```python
#제어문 예제 2
for i in range(1,121):           #1부터 120개의 *을 찍는다.
    print("*",end="")
    if(i%30==0): print(end="\n") #index가 30 나누어떨어지면개행
                                  #따라서 인덱스를 1,121로함
```

```python
# 다중반복구조 예제 1
#구구단 예제
for i in range(1, 10):
    for j in range(1, 10):
        print(i, "*", j, "=", i * j)
    print("********************")
```

```python
# 다중반복구조 예제 1
for i in range(1,10):        #반복문을 1부터 9까지 반복
    for j in range(i,10) :   # i부터 9까지니까 i값이 올라감에 따라 *개수 줄어둔다.
        print("*", end="")
    print(end="\n")
```

```python
# 다중반복구조 예제 2 (while)
i = 1
while True :
    j=i
    while True :
        if(j==10) : break
        print( "*", end="")
        j+=1
    if(i==10) : break
    print(end="\n")
    i+=1
```

```python
#심화실습 1 (달력)
Mon = 1
for LastDay in [31,28,31,30,31,30,31,31,30,31,30,31] :
    print(Mon, "월 : ",end="")
    for Day in range(1, LastDay+1):    #배열의 해당 값이 포함되는 날까지 출력해야돼서 +1
        print(Day, end=",")
    print(end="\n")
    Mon+= 1
```

```python
#심화실습 2 (구구단) - 가로로 출력
for i in range(2,10,3):     #2부터 9까지 3씩 올라간다.
    for j in range(2,10) :  #2~9
        print(i, "*", j,"=",i*j,end="\t")  #1번째 열
        print(i+1, "*", j, "=",i*j,end="\t")  #2번째 열
        if(i+2!=10) : print(i+2,"*",j,"=",i*j,end="\n") #3번째 열
        else : print(end = "\n")
    print(end="\n")
```

```python
#심화실습 3 (구구단) - 세로로 출력
for i in range(2,5):     #세로로 2~4 출력
    for j in range(2,10) :  #2~9
        print(i, "*", j,"=",i*j,end="\t")  #1번째 열
        print(i+3, "*", j, "=",i*j,end="\t")  #2번째 열
        if(i+6!=10) : print(i+6,"*",j,"=",i*j,end="\n") #3번째 열
        else : print(end = "\n")
    print(end="\n")
```
