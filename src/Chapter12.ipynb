```python
#파일 다루기
#파일을 읽고 쓰는 프로그램은 파일 제어를 시스템 운영체계에 의존함
#발생되는 에러에 대해서 예외처리 필수
#시스템에서 에러가 발생하면 except 블록으로 점프하여 처리됨
#에러 발생 및 정상 실행 후 맨 마지막에 finally 블록을 실행

f = 0
try:
    f=open("LICENSE.txt","rt")
    while True:
        r = f.readline()
        if not r:
            break
        print(r)
except FileNotFoundError:  #예외처리(파일이 없을 때 실행된다.)
    print("파일 없음")
finally:               #마지막에 실행되어야하는 부분
    if(f): f.close()
    print("프로그램 끝")
```

```python
#파일 쓰기
#100명의 점수를 임의로 만든 후 csv 형태 파일에 기록하는 예제
import random
f=0
f=open("score.txt","wt")
for i in range(100):
# 아래와 같은 형태를 만들기 위하여 str.format 함수를 사용할 수 있다.
#  one_person = "{0:s},{1:d},{2:d},{3:d}\n"
# .format("홍길동"+str(i), r.randrange(0,100), r.randrange(0,100),r.randrange(0,100))

    one_person = "홍길동"+str(i+1)+","+str(random.randrange(0,100))+","+str(random.randrange(0,100))+","+str(random.randrange(0,100))+"\n"

    f.write(one_person)
    print(one_person)
f.close()
```

```python
#파일 읽기
#앞서 만든 csv 파일을 한 줄씩 읽어서 처리하는 예제

f=0
f=open("score.txt","rt")
while True:
    one_person = f.readline()
    if not one_person : break
    one_person = one_person.replace("\n","")
    item = one_person.split(",")
    print("="*15)
    print("이름: ",item[0])
    print("국어: ",item[1])
    print("영어: ",item[2])
    print("수학: ",item[3])
    print("="*15)
f.close()
```

```python
#데이터 선택
#select에 경우 fetch를 사용하여 데이터를 반복적으로 가져옴
import sqlite3 as s
con = s.connect("examtable.db")
cursor = con.cursor()

cursor.execute("select * from HeyTable")

one_rec = cursor.fetchall()         #해당 데이터 전부 가져옴

for field in one_rec :               #해당 데이터를 반복문을 통해 하나씩 처리
    kor=field[2]
    eng=field[3]
    mat=field[4]
    print("이름:%s 학번:%s 국어:%3d 영어:%3d 수학:%3d 총점:%3d 평균:%3d"%(
field[0],field[1],kor,eng,mat,kor+eng+mat,int((kor+eng+mat)/3)))
    
cursor.close()
con.close()
```
